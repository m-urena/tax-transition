# Bison Wealth Management — Portfolio Transition Optimizer
Finds the optimal set of trades to transition a client portfolio to a target model while minimizing tax impact.
Sweeps across tax budgets to build an efficient frontier showing the tradeoff between tax liability and % transitioned.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.patheffects as pe
from matplotlib.gridspec import GridSpec
from matplotlib.patches import FancyBboxPatch
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from scipy.optimize import linprog
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
plt.rcParams['figure.dpi'] = 130

# ── Bison Brand Colors ─────────────────────────────────────────────────────────
NAVY    = '#1A3A5C'
GOLD    = '#C8A951'
BLUE    = '#2E86C1'
LIGHT   = '#EEF1F5'
WHITE   = '#FFFFFF'
GREEN   = '#1E8449'
RED     = '#C0392B'
GRAY    = '#7F8C8D'
DARK    = '#1C2833'
MUTED   = '#AEB6BF'
CARD_BG = '#F7F9FB'

plt.rcParams.update({
    'font.family': 'DejaVu Sans',
    'figure.facecolor': WHITE,
    'axes.facecolor': WHITE,
    'text.color': DARK,
})

print('Setup complete.')
```

```python
# ── Current Client Holdings ────────────────────────────────────────────────────
# Client: Acme Family Trust  |  Account: *7842  |  As of: March 20, 2026

holdings_data = {
    'Ticker':     ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'NVDA', 'META', 'GE',  'XOM',
                   'JNJ',  'PG',   'TSLA',  'BAC',  'INTC', 'DIS',  'PFE'],
    'Name':       ['Apple Inc', 'Microsoft Corp', 'Alphabet Inc', 'Amazon.com', 'NVIDIA Corp',
                   'Meta Platforms', 'GE Aerospace', 'Exxon Mobil', 'Johnson & Johnson',
                   'Procter & Gamble', 'Tesla Inc', 'Bank of America', 'Intel Corp',
                   'Walt Disney Co', 'Pfizer Inc'],
    'Shares':     [200,   150,   50,    100,   60,    120,   500,   250,
                   200,   150,   80,    400,   300,   200,   400],
    'Price':      [182.0, 378.0, 177.0, 187.0, 812.0, 485.0, 168.0, 108.0,
                   158.0, 157.0, 198.0,  34.0,  29.0,  92.0,  27.0],
    'CostBasis':  [ 45.0, 120.0,  85.0,  95.0, 218.0, 198.0,  85.0,  74.0,
                   145.0, 128.0, 285.0,  42.0,  56.0, 118.0,  41.0],
    'Holding':    ['LT','LT','LT','LT','LT','LT','LT','LT',
                   'LT','LT','ST','LT','LT','LT','LT'],  # LT=long-term, ST=short-term
}

holdings = pd.DataFrame(holdings_data)
holdings['MarketValue']   = holdings['Shares'] * holdings['Price']
holdings['CostValue']     = holdings['Shares'] * holdings['CostBasis']
holdings['UnrealizedGL']  = holdings['MarketValue'] - holdings['CostValue']
holdings['UnrealizedPct'] = holdings['UnrealizedGL'] / holdings['CostValue'] * 100
holdings['TaxRate']       = holdings['Holding'].map({'LT': 0.20, 'ST': 0.37})
# Max tax per position if fully sold
holdings['MaxTax']        = holdings['UnrealizedGL'].clip(lower=0) * holdings['TaxRate']

TOTAL_VALUE = holdings['MarketValue'].sum()
holdings['Weight'] = holdings['MarketValue'] / TOTAL_VALUE

total_gains  = holdings['UnrealizedGL'].clip(lower=0).sum()
total_losses = holdings['UnrealizedGL'].clip(upper=0).sum()
MAX_TAX      = holdings['MaxTax'].sum()

print(f"Portfolio Value      :  ${TOTAL_VALUE:>12,.0f}")
print(f"Gross Unrealized Gain:  ${total_gains:>12,.0f}")
print(f"Gross Unrealized Loss:  ${total_losses:>12,.0f}")
print(f"Max Tax Liability    :  ${MAX_TAX:>12,.0f}")
print()

display_cols = ['Ticker','Shares','Price','CostBasis','MarketValue','UnrealizedGL','UnrealizedPct','Holding']
style = holdings[display_cols].style \
    .format({'Price':'${:.2f}','CostBasis':'${:.2f}','MarketValue':'${:,.0f}',
             'UnrealizedGL':'${:,.0f}','UnrealizedPct':'{:+.1f}%'}) \
    .applymap(lambda v: f'color:{GREEN}' if isinstance(v,(int,float)) and v>0
              else (f'color:{RED}' if isinstance(v,(int,float)) and v<0 else ''),
              subset=['UnrealizedGL','UnrealizedPct']) \
    .set_caption('Current Holdings — Acme Family Trust')
display(style)
```

```python
# ── Target Model Portfolio (20 stocks) ────────────────────────────────────────

model_data = {
    'Ticker': ['AAPL','MSFT','NVDA','GOOGL','AMZN','META','BRK.B','JPM','V','MA',
               'UNH','HD','COST','LLY','ABBV','CRM','NFLX','AMD','NOW','PLTR'],
    'Name':   ['Apple Inc','Microsoft Corp','NVIDIA Corp','Alphabet Inc','Amazon.com',
               'Meta Platforms','Berkshire Hathaway','JPMorgan Chase','Visa Inc',
               'Mastercard Inc','UnitedHealth Group','Home Depot','Costco Wholesale',
               'Eli Lilly','AbbVie Inc','Salesforce Inc','Netflix Inc','AMD',
               'ServiceNow','Palantir Technologies'],
    'TargetWeight': [0.09, 0.09, 0.08, 0.06, 0.06, 0.06, 0.05, 0.05, 0.05, 0.05,
                     0.05, 0.04, 0.04, 0.05, 0.04, 0.04, 0.04, 0.05, 0.04, 0.05],
    'Sector': ['Tech','Tech','Tech','Tech','Consumer','Tech','Financial','Financial',
               'Financial','Financial','Healthcare','Consumer','Consumer','Healthcare',
               'Healthcare','Tech','Consumer','Tech','Tech','Tech'],
}

model = pd.DataFrame(model_data)
model['TargetWeight'] = model['TargetWeight'] / model['TargetWeight'].sum()  # normalize
model['TargetValue']  = model['TargetWeight'] * TOTAL_VALUE

# Overlap analysis
current_set = set(holdings['Ticker'])
target_set  = set(model['Ticker'])
overlap     = current_set & target_set
to_sell     = current_set - target_set
to_buy      = target_set - current_set

print(f"Positions in both (reweight/keep): {len(overlap)} — {sorted(overlap)}")
print(f"Positions to sell (not in model): {len(to_sell)} — {sorted(to_sell)}")
print(f"New positions to buy:             {len(to_buy)} — {sorted(to_buy)}")
print()

display(model[['Ticker','Name','Sector','TargetWeight','TargetValue']].style
    .format({'TargetWeight':'{:.1%}','TargetValue':'${:,.0f}'})
    .set_caption('Target Model Portfolio (20 stocks)'))
```

```python
# ── Tax-Aware Transition Optimizer ────────────────────────────────────────────
#
# Linear Program:
#   Variables : x_i ∈ [0,1]  fraction of position i to liquidate
#   Objective : maximize Σ(x_i · value_i)   [maximize transition proceeds]
#   Constraint: Σ(x_i · gain_i · rate_i) ≤ T   [net tax ≤ budget]
#
# Key insight: loss positions have negative tax_per_unit, so selling them
# creates tax credits that offset gains — the optimizer harvests losses first.

def solve_transition(holdings_df, tax_budget):
    """Return optimal sell fractions and summary metrics for a given tax budget."""
    n          = len(holdings_df)
    values     = holdings_df['MarketValue'].values
    gains      = holdings_df['UnrealizedGL'].values
    rates      = holdings_df['TaxRate'].values
    tax_coeffs = gains * rates   # negative for loss positions

    result = linprog(
        c      = -values,                     # minimize negative = maximize proceeds
        A_ub   = tax_coeffs.reshape(1, -1),   # net tax ≤ budget
        b_ub   = np.array([tax_budget]),
        bounds = [(0, 1)] * n,
        method = 'highs',
    )

    if not result.success:
        return None

    x            = result.x
    proceeds     = float(values @ x)
    realized_gl  = float(gains @ x)
    net_tax      = float(tax_coeffs @ x)
    realized_tax = max(0.0, net_tax)   # floor: can't have negative tax bill

    return {
        'x':              x,
        'proceeds':       proceeds,
        'realized_gl':    realized_gl,
        'net_tax':        realized_tax,
        'transition_pct': proceeds / holdings_df['MarketValue'].sum() * 100,
        'trades':         holdings_df['Ticker'].values,
        'sell_fracs':     x,
    }


# Sanity check
_r0   = solve_transition(holdings, 0)
_rmax = solve_transition(holdings, MAX_TAX)
print(f"T=$0       → {_r0['transition_pct']:.1f}% transitioned  |  Tax: ${_r0['net_tax']:,.0f}  |  G/L: ${_r0['realized_gl']:,.0f}")
print(f"T=max      → {_rmax['transition_pct']:.1f}% transitioned  |  Tax: ${_rmax['net_tax']:,.0f}  |  G/L: ${_rmax['realized_gl']:,.0f}")
```

```python
# ── Efficient Frontier Sweep ───────────────────────────────────────────────────

N_POINTS    = 300
tax_budgets = np.linspace(0, MAX_TAX * 1.02, N_POINTS)
frontier    = []

for T in tax_budgets:
    r = solve_transition(holdings, T)
    if r:
        frontier.append({
            'tax_budget':     T,
            'tax_incurred':   r['net_tax'],
            'transition_pct': r['transition_pct'],
            'realized_gl':    r['realized_gl'],
            'proceeds':       r['proceeds'],
        })

frontier = pd.DataFrame(frontier).drop_duplicates(subset='transition_pct').reset_index(drop=True)
print(f"Efficient frontier: {len(frontier)} unique points computed.")
print(f"  Min transition: {frontier['transition_pct'].min():.1f}%  →  "
      f"Max transition: {frontier['transition_pct'].max():.1f}%")
print(f"  Tax range: ${frontier['tax_incurred'].min():,.0f} – ${frontier['tax_incurred'].max():,.0f}")
```

```python
# ── Scenario Definitions ──────────────────────────────────────────────────────

CUSTOM_BUDGET = MAX_TAX * 0.52   # ← adjust this to change the Custom scenario

scenario_defs = [
    dict(name='Minimum Tax',    budget=0,               color=GREEN, label_color='#145A32'),
    dict(name='Moderate Tax',   budget=MAX_TAX * 0.28,  color=GOLD,  label_color='#7D6608'),
    dict(name='Max Transition', budget=MAX_TAX,          color=RED,   label_color='#922B21'),
    dict(name='Custom',         budget=CUSTOM_BUDGET,    color=BLUE,  label_color='#1A5276'),
]

scenarios = []
for d in scenario_defs:
    r = solve_transition(holdings, d['budget'])
    scenarios.append({**d, **r})
    print(f"{d['name']:<16}  "
          f"transition={r['transition_pct']:5.1f}%  "
          f"tax=${r['net_tax']:>9,.0f}  "
          f"G/L=${r['realized_gl']:>10,.0f}")
```

```python
# ── Helper: draw donut ─────────────────────────────────────────────────────────

def draw_donut(ax, pct, color, label='% Managed'):
    """Draw a donut chart showing pct% in color on a muted background ring."""
    bg   = '#DDE3EA'
    frac = np.clip(pct / 100, 0, 1)
    # Outer background ring
    ax.pie([1], colors=[bg], startangle=90,
           wedgeprops=dict(width=0.34, edgecolor=WHITE, linewidth=2.5))
    # Filled arc (transition amount), then remainder
    if frac > 0:
        ax.pie([frac, 1 - frac], colors=[color, bg], startangle=90,
               wedgeprops=dict(width=0.34, edgecolor=WHITE, linewidth=2.5))
    # Center labels
    ax.text(0,  0.10, f'{pct:.0f}%',  ha='center', va='center',
            fontsize=19, fontweight='bold', color=DARK)
    ax.text(0, -0.22, label,          ha='center', va='center',
            fontsize=8,  color=GRAY)
    ax.set_aspect('equal')
    ax.axis('off')


def fmt_dollar(v, show_sign=False):
    """Format a number as a dollar amount."""
    prefix = '+' if (show_sign and v > 0) else ('−' if v < 0 else '')
    return f'{prefix}${abs(v):,.0f}'


def metric_row(ax, y, label, value_str, value_color=DARK, label_size=9, val_size=16):
    """Draw a metric label + value pair at vertical position y (axes coords 0–1)."""
    ax.text(0.5, y + 0.035, label, ha='center', va='bottom',
            fontsize=label_size, color=GRAY, transform=ax.transAxes)
    ax.text(0.5, y - 0.008, value_str, ha='center', va='top',
            fontsize=val_size, fontweight='bold', color=value_color,
            transform=ax.transAxes)


print('Helper functions ready.')
```

```python
# ── Main Dashboard ─────────────────────────────────────────────────────────────

FIG_W, FIG_H = 16, 8.5
fig = plt.figure(figsize=(FIG_W, FIG_H), facecolor=LIGHT)

# ── Header strip ──────────────────────────────────────────────────────────────
ax_hdr = fig.add_axes([0.0, 0.90, 1.0, 0.10])
ax_hdr.set_facecolor(NAVY)
ax_hdr.axis('off')

ax_hdr.text(0.018, 0.72, 'BISON', fontsize=20, fontweight='bold',
            color=GOLD, va='top', transform=ax_hdr.transAxes,
            fontfamily='DejaVu Sans')
ax_hdr.text(0.018, 0.22, 'WEALTH MANAGEMENT', fontsize=7.5, color=MUTED,
            va='top', transform=ax_hdr.transAxes, fontweight='bold')

ax_hdr.text(0.5, 0.62, 'PORTFOLIO TRANSITION OPTIMIZER',
            fontsize=15, fontweight='bold', color=WHITE,
            ha='center', va='center', transform=ax_hdr.transAxes)
ax_hdr.text(0.5, 0.20, 'Tax-Efficient Rebalancing Analysis  |  Tax Liability vs. Transition Trade-off',
            fontsize=8, color=MUTED, ha='center', va='center', transform=ax_hdr.transAxes)

ax_hdr.text(0.982, 0.68, 'Acme Family Trust   |   Acct *7842',
            fontsize=7.5, color=MUTED, ha='right', va='center', transform=ax_hdr.transAxes)
ax_hdr.text(0.982, 0.28, f'Portfolio Value: ${TOTAL_VALUE:,.0f}   |   March 20, 2026',
            fontsize=7,   color=MUTED, ha='right', va='center', transform=ax_hdr.transAxes)

ax_hdr.plot([0, 1], [0, 0], color=GOLD, linewidth=2, transform=ax_hdr.transAxes)


# ── Scenario columns ──────────────────────────────────────────────────────────
N      = len(scenarios)
CARD_L = 0.012
CARD_B = 0.030
CARD_H = 0.855
GAP    = 0.010
CARD_W = (1.0 - 2 * CARD_L - (N - 1) * GAP) / N

def hdivider(ax, y, xmin=0.08, xmax=0.92):
    ax.plot([xmin, xmax], [y, y], color='#D5DADE', linewidth=0.7,
            transform=ax.transAxes, clip_on=False)

for col_i, sc in enumerate(scenarios):
    left    = CARD_L + col_i * (CARD_W + GAP)
    card_ax = fig.add_axes([left, CARD_B, CARD_W, CARD_H])
    card_ax.set_facecolor(CARD_BG)
    card_ax.axis('off')
    for spine in card_ax.spines.values():
        spine.set_visible(False)

    # Colored scenario header strip
    header_h = 0.10
    strip = FancyBboxPatch((0, 1 - header_h), 1, header_h,
                           boxstyle='square,pad=0', linewidth=0,
                           facecolor=sc['color'], transform=card_ax.transAxes,
                           clip_on=False)
    card_ax.add_patch(strip)
    card_ax.text(0.5, 1 - header_h / 2, sc['name'].upper(),
                 ha='center', va='center', fontsize=10, fontweight='bold',
                 color=WHITE, transform=card_ax.transAxes)

    # ── Metrics ───────────────────────────────────────────────────────────────
    tx_color = RED   if sc['net_tax']    > 0  else GREEN
    gl_color = GREEN if sc['realized_gl'] <= 0 else RED

    metric_row(card_ax, 0.760, 'ESTIMATED TAX LIABILITY',
               fmt_dollar(sc['net_tax']), tx_color, 7, 15)
    hdivider(card_ax, 0.715)

    metric_row(card_ax, 0.635, 'REALIZED GAIN / LOSS',
               fmt_dollar(sc['realized_gl'], show_sign=True), gl_color, 7, 15)
    hdivider(card_ax, 0.590)

    metric_row(card_ax, 0.510, 'TRANSITION AMOUNT',
               fmt_dollar(sc['proceeds']), NAVY, 7, 15)
    card_ax.text(0.5, 0.462, f"{sc['transition_pct']:.1f}% of portfolio",
                 ha='center', va='top', fontsize=8, color=GRAY,
                 transform=card_ax.transAxes)
    hdivider(card_ax, 0.425)

    # ── Donut chart ───────────────────────────────────────────────────────────
    iax = inset_axes(card_ax, width='68%', height='36%',
                     loc='lower center',
                     bbox_to_anchor=(0.5, 0.02, 1, 1),
                     bbox_transform=card_ax.transAxes,
                     borderpad=0)
    draw_donut(iax, sc['transition_pct'], sc['color'])

    # Tax budget footnote
    budget_label = 'Tax Budget: $0' if sc['budget'] == 0 else f"Tax Budget: ${sc['budget']:,.0f}"
    card_ax.text(0.5, 0.022, budget_label,
                 ha='center', va='bottom', fontsize=7, color=GRAY,
                 style='italic', transform=card_ax.transAxes)

    # Card border
    for side in ['top', 'bottom', 'left', 'right']:
        card_ax.spines[side].set_visible(True)
        card_ax.spines[side].set_color('#C8D0D8')
        card_ax.spines[side].set_linewidth(0.7)


# ── Footer ────────────────────────────────────────────────────────────────────
fig.text(0.5, 0.005,
         'Illustrative purposes only. Tax estimates assume LT rate 20%, ST rate 37%. '
         'Actual liability may differ. Consult a qualified tax advisor before transacting.',
         ha='center', fontsize=6.5, color=GRAY, style='italic')

plt.show()
```

```python
# ── Trade-Level Detail (per scenario) ─────────────────────────────────────────

def build_trade_table(sc):
    """Return a DataFrame of recommended trades for a given scenario result."""
    rows = []
    for i, ticker in enumerate(holdings['Ticker']):
        frac = sc['sell_fracs'][i]
        if frac < 0.005:   # skip negligible
            continue
        row = holdings.iloc[i]
        shares_sold    = frac * row['Shares']
        proceeds       = shares_sold * row['Price']
        realized_gl    = shares_sold * (row['Price'] - row['CostBasis'])
        tax_impact     = max(0, realized_gl) * row['TaxRate']
        rows.append({
            'Ticker':       ticker,
            'Action':       'SELL',
            'Sell %':       frac * 100,
            'Shares Sold':  shares_sold,
            'Proceeds':     proceeds,
            'Realized G/L': realized_gl,
            'Tax Impact':   tax_impact,
            'Holding':      row['Holding'],
        })
    return pd.DataFrame(rows)


for sc in scenarios:
    tbl = build_trade_table(sc)
    if tbl.empty:
        print(f"\n{sc['name']}: No trades.")
        continue
    print(f"\n{'─'*60}")
    print(f"  {sc['name'].upper()}  |  {len(tbl)} sell trades  |  "
          f"Total proceeds: ${tbl['Proceeds'].sum():,.0f}  |  "
          f"Total tax: ${tbl['Tax Impact'].sum():,.0f}")
    print(f"{'─'*60}")
    display(tbl.style
        .format({'Sell %':'{:.0f}%','Shares Sold':'{:.1f}',
                 'Proceeds':'${:,.0f}','Realized G/L':'${:,.0f}','Tax Impact':'${:,.0f}'})
        .applymap(lambda v: f'color:{GREEN}' if isinstance(v,(int,float)) and v<0
                  else (f'color:{RED}' if isinstance(v,(int,float)) and v>0 else ''),
                  subset=['Realized G/L','Tax Impact'])
        .hide(axis='index'))
```
